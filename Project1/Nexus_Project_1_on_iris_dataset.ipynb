## Importing all the required libraries for Analysizing tghe Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Data Collection

```python
#Import the csv file from internal drive to the current notebook.
#Create a dataframe for the imported data.
iris_df = pd.read_csv(r"C:\Users\Ranjith\Downloads\Iris (1).csv")
```

```python
iris_df.drop(columns = ['Id'],inplace=True)
```

```python
iris_df.head()
```

```python
# visualizing the raw dataset.
iris_df.info()
```

```python
# Checking the basic statistics of the data.
iris_df.describe()
```

```python
# checking for any null values in the dataframe.
iris_df.isnull().sum()
```

```python
#Checking for any duplicate values in the dataset.
iris_df.duplicated().sum()
```

```python
iris_df.drop_duplicates(inplace=True)
```

## Pair-wise Relationships:

```python
# Pairplot for pairwise relationships
sns.pairplot(iris_df, hue='Species')
plt.show()
```

#### The pair plot revealed distinct patterns, especially the clear separation between the 'setosa' species and the other two species in various feature combinations.

```python
# Boxplot for feature distribution
# Using box plot to check for any outliers in the data
plt.figure(figsize=(12, 6))
sns.boxplot(x='Species', y='SepalLengthCm', data=iris_df)
plt.title('Distribution of Sepal Length by Species')
plt.show()
```

```python
# Count plot representing count of each category in the Species column.
sns.countplot(data = iris_df,x = 'Species')
plt.xlabel("Type of the Species")
plt.ylabel("Count of the Species")
plt.show()
```

## Correlation Analysis:

```python
correlation_matrix = iris_df.corr()
# Create a heatmap using Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

### The correlation matrix and heatmap indicated potential correlations between features. 
### Notably, a strong positive correlation was observed between 'petal_length' and 'petal_width'.

## Distribution Analysis:

```python
sns.histplot(data=iris_df, x='SepalLengthCm', hue='Species', kde=True)
plt.show()
```

```python
sns.histplot(data=iris_df, x='SepalWidthCm', hue='Species', kde=True)
plt.show()
```

```python
sns.histplot(data=iris_df, x='PetalLengthCm', hue='Species', kde=True, bins=20)
plt.show()
```

```python
sns.histplot(data=iris_df, x='PetalWidthCm', hue='Species', kde=True, bins=20)
plt.show()
```

#### Histograms showed clear differences in feature distributions for different iris species, providing valuable insights into how features vary across species.
