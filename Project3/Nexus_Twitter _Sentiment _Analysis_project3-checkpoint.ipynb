### Text Preprocessing:
- Preprocess tweet text by removing stop words, special characters, and URLs.
- Tokenize and lemmatize words to prepare the text for sentiment analysis.

```python
df.text
```

```python
#Checking whether the given data is in correct text formate
def edat(data,name):
    case=' '.join(data[name]).islower()
    html = data[name].apply(lambda x:True if re.findall("<.+?>",x)else False).sum()
    urls = data[name].apply(lambda x:True if re.findall("http[s]?://.+? +",x)else False).sum()
    unwanted_ = data[name].apply(lambda x:True if re.findall("[]()*\-.,@#$%^&0-9]",x)else False).sum()
    if case==False:
        print('not in lower case')
    if html>0:
        print('have html tags')
    if urls>0:
        print('have urls ')
    if unwanted_>0:
        print('having unwanted char')
    else:
        print("ntg")
```

```python
edat(df,'text')
```

```python
fv = df["text"]
fv
```

```python
def textpp(x,correct,emojii):
    from textblob import TextBlob
    x = x.lower()
    if emojii == True:
        x=emoji.demojize(x)
    x = re.sub("<.+?>"," ",x)
    x = re.sub("[.+]?http[s]?://.+? +"," ",x)
    x = re.sub("[]:()*\-.,;@#$%^&0-9!\?]"," ",x)
    if correct == 't':
        x = TextBlob(x).correct().string
    else:
        return x
    return x
```

```python
fv=fv.apply(textpp,args=("s",False))
```

```python
fv
```

```python
stp = stopwords.words("english")
```

```python
def advpp(x,stemm):
    wl = WordNetLemmatizer()
    l = []
    for word in word_tokenize(x):
        if word in stp:
            pass
        else:
            if stemm == 'r':
                l.append(wl.lemmatize(word,pos = 'v'))
            else:
                l.append(word)      
    return ' '.join(l)
```

```python
fv
```

```python
fv=fv.apply(advpp,args=('r'))
```

```python
#Model1
cv1=CountVectorizer()
```
